```python
import matplotlib.pyplot as plt
import numpy as np
X = ["Day -28 to -2\n(Screening)", "Day -1\n(Check-In)"]+["Day "+str(i*7) for i in range(1,4)]
y1 = [7,4.26,3.33,3.49,3.33]
y2 = [0.79,0.79,0.94,0.91,0.81]

fig = plt.figure(figsize=(20,5))
fig.set_facecolor("white")
ax1 = plt.subplot(1,2,1)
plt.title("White Blood Cell", fontsize=18)
plt.xlabel("Visit", fontsize=18)
plt.ylabel("Count (x10^3/uL)", fontsize=18)
plt.ylim(min(y1)-0.5, max(y1)+0.5)
plt.plot(y1, marker="o", color="blue")
for xy in zip(range(5),y1):
    xy = tuple([xy[0], xy[1]+0.25])
    ax1.annotate(xy[1], xy=xy, textcoords='data')
plt.xticks(np.arange(5), ["Day -28 to -2\n(Screening)", "Day -1\n(Check-In)"]\
    +["Day "+str(i*7) for i in range(1,4)], fontsize=12)

ax2 = plt.subplot(1,2,2)
plt.plot(y2, marker="o")
plt.title("Serum Creatinine", fontsize=18)
plt.xlabel("Visit", fontsize=18)
plt.ylabel("Concentration (mg/dL)", fontsize=18)
plt.ylim(min(y2)-0.05, max(y2)+0.05)
plt.plot(y2, marker="o", color="blue")
for xy in zip(range(5),y2):
    xy = tuple([xy[0], round(xy[1]+0.01,2)])
    if xy[0]==1:
        xy = tuple([xy[0]-0.1, round(xy[1],2)])
    ax2.annotate(xy[1], xy=xy, textcoords='data')
plt.xticks(np.arange(5), ["Day -28 to -2\n(Screening)", "Day -1\n(Check-In)"]\
    +["Day "+str(i*7) for i in range(1,4)], fontsize=12)

plt.savefig("random.jpg")
#plt.show()
```
