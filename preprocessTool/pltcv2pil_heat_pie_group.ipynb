```python
import cv2
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
from PIL import Image

mp = mpimg.imread('groundhog.png')
cv = cv2.imread('groundhog.png')
pl = Image.open("groundhog.png")

print(type(mp), type(cv), type(pl))
print(mp.shape, cv.shape, pl.size)
```

```python
plt.imshow(mp)
cv2.imshow("A", cv); cv2.waitKey(0); cv2.destroyAllWindows()
pl.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# heat map # plt.imshow # plt.colorbar
data = np.zeros((100,100))
for i in range(len(data)):
    for j in range(len(data[0])):
        data[i][j] = (0.01*i)**2+(0.01*j)**2
plt.imshow(data)
plt.colorbar()
plt.show()
```

```python
# heatmap with value # ax.text
data = np.random.random((10, 15))
fig, ax = plt.subplots()
fig.set_facecolor('green')
for i in range(len(data)):
    for j in range(len(data[0])):
        ax.text(j, i, round(data[i][j],1), ha="center", va="center", color="b")
plt.imshow(data, cmap='hot')
plt.colorbar()
plt.show()
```

```python
# pie
L = [0.3, 0.2, 0.5]
plt.pie(L, labels=["blue", "orange", "green"], autopct="%.1f%%")
plt.show()
```

```python
# group bar
students = ['Jack', 'Mary', 'Mike', 'David']
math = [78, 67, 90, 81]
history = [94, 71, 65, 88]
R, width = np.arange(len(students)), 0.3
plt.bar(R, math, width, color='green')
plt.bar(R+width, history, width, color='blue')
plt.xticks(R+width/2, students)
plt.show()
```
