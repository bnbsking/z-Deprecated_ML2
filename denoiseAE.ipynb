```python
from keras.datasets import mnist
from keras.layers import Input, Dense, Dropout
from keras.models import Model
from keras.optimizers import Adam
from keras.losses import mse
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
(xTrain0, _), (xVal0, _) = mnist.load_data()
xTrain0, xVal0 = xTrain0.reshape(60000,784)/255., xVal0.reshape(10000,784)/255.

xTrain, xVal = xTrain0.copy(), xVal0.copy()
for i in range(60000):
    xTrain[i]+=0.3*np.random.randn(784)
for i in range(10000):
    xVal[i]+=0.3*np.random.randn(784)
```

```python
inputL = Input(shape=(784,))
x = Dense(512, activation='relu')(inputL)
x = Dropout(0.1)(x)
x = Dense(64, activation='relu')(x)
x = Dropout(0.1)(x)
x = Dense(10, activation='relu')(x)
x = Dense(64, activation='relu')(x)
x = Dropout(0.1)(x)
x = Dense(512, activation='relu')(x)
x = Dropout(0.1)(x)
outputL = Dense(784, activation='relu')(x)
model = Model(inputL, outputL)
model.summary()
model.compile(optimizer=Adam(lr=0.0001), loss=mse)
```

```python
history = model.fit(xTrain, xTrain0, validation_data=(xVal, xVal0), epochs=300, batch_size=10000)
```

```python
plt.figure(figsize=(8,5))
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.legend(labels=['loss', 'val_loss'])
plt.show()#150
```

```python
pred = model.predict(xTrain)
```
